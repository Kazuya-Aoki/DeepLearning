```python
import numpy as np

def init_network():
    network = {}
    network['W1'] = np.array([[0.1,0.3,0.5], [0.2,0.4,0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1,0.4], [0.2,0.5], [0.3,0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1,0.3], [0.2,0.4]])
    network['b3'] = np.array([0.1, 0.2])
    
    return network


def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = identity_function(a3)
    
    return y


def identity_function(x):
    return x


def sigmoid(a):
    return 1 / (1+np.exp(-a))


nw = init_network()
x = np.array([1.0, 0.5])
y = forward(nw, x)
print(y)
```

```python
import numpy as np

def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    
    y = exp_a / sum_exp_a
    
    return y

in_a = np.array([23, 11, 81, 50])
out_a = softmax(in_a)

print("Softmax Function: {}\nOutput\'s sum: {}".format(out_a, sum(out_a)))
```

```python
from Function.Fn import get_data as gt
from Function.Fn import predict
import numpy, pickle

print("")

def init_network():
    with open("sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)
    return network

x, t = gt()
network = init_network()

batch_size = 100 #
accuracy_cnt = 0

for i in range(0, len(x), batch_size):
    x_batch = x[i: i+batch_size]
    y_batch = predict(network, x_batch)
    p = numpy.argmax(y_batch, axis=1)
    accuracy_cnt += numpy.sum(p == t[i:i+batch_size])

print("Accuracy: {}".format(str(float(accuracy_cnt)/len(x))))    
```
